```python
w_think_m=woman.drop(['attr1_2','sinc1_2','intel1_2','fun1_2','amb1_2','shar1_2','gender'],axis=1)
w_think_m=w_think_m.dropna()

m_think_w=man.drop(['attr1_2','sinc1_2','intel1_2','fun1_2','amb1_2','shar1_2','gender'],axis=1)
m_think_w=m_think_w.dropna()

man_want=man.drop(['attr2_2','sinc2_2','intel2_2','fun2_2','amb2_2','shar2_2','gender'],axis=1)
man_want=man_want.dropna()

woman_want=woman.drop(['attr2_2','sinc2_2','intel2_2','fun2_2','amb2_2','shar2_2','gender'],axis=1)
woman_want=woman_want.dropna()
```

```python
B=w_think_m.mean()
w_t=list()
for i in B:
    i=int(i)
    w_t.append(i)
w_t.append(50)
w_t
```

```python
import networkx as nx
%matplotlib inline
Women_per = nx.Graph()
for i in range(0,len(feature)-1):
    Women_per.add_node(feature[i],pos=(5,w_t[i]*100))
    Women_per.add_node(feature_1[i],pos=(15,m_w[i]*100))
    Women_per.add_edge(feature[i],feature_1[i])
Women_per.add_node(feature[6],pos=(5,5000))
Women_per.add_node(feature_1[6],pos=(15,5000))
nx.draw(Women_per, nx.get_node_attributes(Women_per, 'pos'), with_labels=True, node_size=0)
```
