```python
import csv
with open('Speed Dating Data.csv', newline='') as f:
    reader = csv.reader(f)
    rownum=0
    iid_list=[]
    like_o_list=[]
    for row in reader:
        if rownum != 0:  
            iid = int(row[0])
            if row[30]=='':
                row[30] = 0
            like_o = float(row[30])
            iid_list.append(iid)
            like_o_list.append(like_o)
        rownum += 1
    print(iid_list, like_o_list)
        
```

```python

new_like_o_list=[]
like_o_total=0
like_o_count=0
for i in range(len(iid_list)):
    if i == 0:
        like_o_total += like_o_list[i]
        like_o_count = 1
    else:
        if iid_list[i] == iid_list[i-1]:  
            like_o_total += like_o_list[i]
            like_o_count += 1
        else:
            new_like_o_list.append(round(like_o_total/like_o_count,2))
            like_o_total = like_o_list[i]
            like_o_count = 1
    if i == len(iid_list)-1:
        new_like_o_list.append(round(like_o_total/like_o_count,2))
print(new_like_o_list) 
            

```

```python
new_iid_list=[]
new_iid=0
for i in range(len(iid_list)):
    if i == 0:
        new_iid = 1
    else:
        if iid_list[i] == iid_list[i-1]:  
            new_iid = iid_list[i]
        else:
            new_iid_list.append(new_iid)
            new_iid = iid_list[i]
    if i == len(iid_list)-1:
        new_iid_list.append(new_iid)
print(new_iid_list) 
```

```python
new_dict={k:v for k, v in zip(new_iid_list,new_like_o_list)}
new_dict
```

```python
hobbies_texts = ['whobbies','mhobbies']
all_texts = dict()
for text in hobbies_texts:
    with open(text,'r') as f:
        data = f.read()
        all_texts[text] = data
```

```python
%matplotlib inline
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from scipy.misc import imread
import random

def pink_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "#FF66CC"

Amask=imread('woman3.png',flatten=True)
wordcloud = WordCloud(font_path='/Library/Fonts/CabinSketch-Bold.ttf',stopwords=STOPWORDS,width=10000,height=30000,mask=Amask,prefer_horizontal=1,background_color='#FBF7F8').generate(all_texts['whobbies'])


plt.imshow(wordcloud.recolor(color_func=pink_color_func, random_state=3))
plt.axis('off')
plt.savefig('./woman_3.png', dpi=300)
plt.show()
```
