```python
import pandas as pd
import numpy as np
df=pd.read_csv('Feature.csv')
man=df[df['gender']==1].drop(['expnum','gender','race'],axis=1)
woman=df[df['gender']==0].drop(['expnum','gender','race'],axis=1)
wdata=woman.dropna()
mdata=man.dropna()
```

```python
mdata['pid']=mdata['iid']
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv('Speed Dating data.csv')
match=df[df['gender']==0][['iid','gender','pid','match']]
```

```python
match
```

```python
#merge three sheets, to get a sheet with match and attributes of man and woman
wm=match.merge(wdata,on='iid')
fdata=wm.merge(mdata,on='pid')
fdata
```

```python
feature=list(wdata.columns.values)
wfeature=list()
mfeature=list()
for i in feature:
    wfeature.append(i+'_x')
    mfeature.append(i+'_y')
wfeature
```

```python
y=fdata['match']
```

```python
# preprocessing
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(fdata)
fdata=scaler.transform(fdata)
forreg=pd.DataFrame(fdata)
forreg.to_csv('forreg.csv')
```

```python
forreg=pd.read_csv('forreg.csv')
```

```python
#get the difference of attributes of man and woman
for j in range(len(feature)):
    forreg[feature[j]]=abs(forreg[wfeature[j]]-forreg[mfeature[j]])

forreg.to_csv('forreg1.csv')
```
